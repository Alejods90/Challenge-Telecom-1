```python
# Create a copy of the DataFrame after filtering to avoid SettingWithCopyWarning
df = df.copy()

new_columns = {}
for col in df.columns:
    if '.' in col:
        parts = col.split('.')
        if len(parts) > 1:
            new_name = parts[-1].capitalize()
            new_columns[col] = new_name

df.rename(columns=new_columns, inplace=True)
display(df.head())
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        # Check if the column contains only 'Yes' and 'No' or values that can be coerced to numeric
        if df[col].isin(['Yes', 'No']).all():
            df[col] = df[col].replace({'Yes': 1, 'No': 0}).astype(int)
        # Handle other object columns that might contain numeric-like strings
        elif df[col].str.isnumeric().all():
             df[col] = pd.to_numeric(df[col])


display(df.head())
```

```python
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')
display(df.info())
```

## Analyze categorical features

### Subtask:
Create visualizations (e.g., bar plots) to see how churn relates to categorical features like 'Gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', and 'PaymentMethod'.


**Reasoning**:
Create bar plots for each categorical feature to visualize the relationship with churn.



```python
categorical_cols = ['Gender', 'Partner', 'Dependents', 'Phoneservice', 'Multiplelines', 'Internetservice', 'Onlinesecurity', 'Onlinebackup', 'Deviceprotection', 'Techsupport', 'Streamingtv', 'Streamingmovies', 'Contract', 'Paperlessbilling', 'Paymentmethod']

for col in categorical_cols:
    churn_by_cat = df.groupby([col, 'Churn']).size().reset_index(name='Count')
    # Convert Churn to string for discrete colors in legend
    churn_by_cat['Churn'] = churn_by_cat['Churn'].astype(str)
    fig = px.bar(churn_by_cat, x=col, y='Count', color='Churn', title=f'Churn Distribution by {col}', color_discrete_sequence=px.colors.qualitative.Plotly)
    fig.show()
```

## Resumen:

### Hallazgos Clave del Análisis de Datos

* Entre las características categóricas, las siguientes muestran diferencias notables en la distribución de la **cancelación de clientes** (churn):
    * Los clientes con **'Servicio de Internet'** diferente de 'No' (DSL o Fibra óptica) tienen una tasa de cancelación más alta.
    * Los clientes sin **'Seguridad en Línea'**, **'Copia de Seguridad en Línea'**, **'Protección de Dispositivos'** o **'Soporte Técnico'** tienden a cancelar más.
    * Los clientes con un contrato **'Mes a mes'** cancelan significativamente más que aquellos con contratos a más largo plazo.
    * Los clientes que usan **'Facturación sin Papel'** tienen una tasa de cancelación más alta.
    * El **'Método de Pago'** también muestra variaciones en las tasas de cancelación entre los diferentes métodos.
* Para las características numéricas:
    * Los clientes que cancelaron ('Sí') tienen una **'Antigüedad'** menor en comparación con aquellos que no cancelaron ('No').
    * Los clientes que cancelaron ('Sí') tienden a tener cargos **'Mensuales'** más altos.
    * Los clientes que no cancelaron ('No') generalmente tienen cargos **'Totales'** más altos, probablemente debido a una mayor antigüedad.
* Con respecto a la característica **'Adulto Mayor'**:
    * No adultos mayores: 4508 no cancelaron, 1393 cancelaron.
    * Adultos mayores: 666 no cancelaron, 476 cancelaron.
    * Los adultos mayores tienen una tasa de cancelación más alta (aproximadamente 41.6%) en comparación con los no adultos mayores (aproximadamente 23.6%).
